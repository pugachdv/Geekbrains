В принципе если есть фичи или возможность сгенерировать дополнительные признаки, то проверка результативности гибридной по сравнению с простой системой будет не лишней.

Использовать следует, где может быть региональная специфика, гендерная, возрастная для юзеров. Для товаров, например, тематическая/жанровая (в контенте).
